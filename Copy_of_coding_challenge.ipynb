```python
## DAY 07 CODING CHALLENGE

print("Welcome to the Shopping Cart System")
print("Enter item prices one by one")
print("Enter 0 to finish billing\n")

item_count = 0
subtotal = 0.0

while True:
    price = float(input("Enter item price: "))

    # Sentinel value to stop input
    if price == 0:
        break

    # Negative price check
    if price < 0:
        print("Price cannot be negative. Please enter again.")
        continue


    item_count += 1
    subtotal += price

else:

    print("Input completed")

# Discount calculation
if subtotal >= 1000:
    discount_rate = 0.20
elif subtotal >= 500:
    discount_rate = 0.10
elif subtotal >= 200:
    discount_rate = 0.05
else:
    discount_rate = 0.0

discount_amount = subtotal * discount_rate
final_total = subtotal - discount_amount

# Bill Summary
print("\n BILL SUMMARY")
print("---------------------------")
print("Total items   :", item_count)
print("Subtotal      : ₹", subtotal)
print("Discount      : ₹", discount_amount)
print("Final Total   : ₹", final_total)
print("---------------------------")
print("Thank you for shopping!")
```

```python
## DAY 08 CODING CHALLENGE

import random


scores = [random.randint(0, 100) for _ in range(10)]

print(" Student Exam Scores")

for index, score in enumerate(scores):
    print(f"Student {index}: Score = {score}")

high_scores = [score for score in scores if score > 75]

average_score = sum(scores) / len(scores)

print("\n Scores greater than 75:")
print(high_scores)

print("\n Average Score:", round(average_score, 2))
```

```python
##  DAY 09 CODING CHALLENGE

def add(a, b):
    return a + b

def subtract(a, b):
    return a - b

def multiply(a, b):
    return a * b

def divide(a, b):
    if b == 0:
        return "Error: Division by zero is not allowed"
    return a / b


# User input
num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))
operation = input("Enter operation (+, -, *, /): ")

# Perform operation
if operation == "+":
    result = add(num1, num2)
elif operation == "-":
    result = subtract(num1, num2)
elif operation == "*":
    result = multiply(num1, num2)
elif operation == "/":
    result = divide(num1, num2)
else:
    result = "Invalid operation"

if isinstance(result, (int, float)):
    print("Result:", round(result, 2))
    print("Absolute Value:", abs(result))
    print("Power (result²):", pow(result, 2))
else:
    print(result)
```

```python
## DAY 10 CODING CHALLENGE

# Global variable
transaction_count = 0

class BankAccount:
    def __init__(self, account_holder, balance=0):
        self.account_holder = account_holder
        self.balance = balance

    def deposit(self, amount):
        global transaction_count
        self.balance += amount
        transaction_count += 1
        print(f"Deposited ₹{amount}")

    def withdraw(self, amount):
        global transaction_count
        if amount > self.balance:
            print("❌ Withdrawal denied: Insufficient balance")
        else:
            self.balance -= amount
            transaction_count += 1
            print(f"Withdrawn ₹{amount}")

    def display_balance(self):
        print("\nAccount Holder:", self.account_holder)
        print("Current Balance: ₹", self.balance)


# Creating object
account1 = BankAccount("Hari", balance=5000)

# Calling methods using positional arguments
account1.deposit(2000)

# Calling methods using keyword arguments
account1.withdraw(amount=3000)

# Trying to withdraw more than balance
account1.withdraw(amount=5000)

# Display balance
account1.display_balance()

# Display global transaction count
print("\nTotal Transactions:", transaction_count)
```
